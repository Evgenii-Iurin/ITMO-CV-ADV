```python
class HandKeypointDetectionDataset(Dataset):
    def __init__(self, image_dir, label_dir, image_size=256):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.image_files = sorted([f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))])
        self.image_size = image_size

        self.transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
        ])

        # Pre-filter valid samples
        self.valid_image_files = self._filter_valid_samples()

    def _filter_valid_samples(self):
        valid_files = []
        for image_name in self.image_files:
            label_path = os.path.join(self.label_dir, image_name.replace('.jpg', '.txt').replace('.png', '.txt'))
            try:
                with open(label_path, 'r') as f:
                    line = f.readline().strip().split()
                    line = list(map(float, line))
                    keypoints = torch.tensor(line[5:], dtype=torch.float32).view(21, 3)
                    if torch.all((keypoints[:, 0] >= 0) & (keypoints[:, 0] <= 1) &
                                 (keypoints[:, 1] >= 0) & (keypoints[:, 1] <= 1)):
                        valid_files.append(image_name)
            except:
                continue  # skip unreadable or malformed files
        return valid_files

    def __len__(self):
        return len(self.valid_image_files)

    def __getitem__(self, idx):
        image_name = self.valid_image_files[idx]
        image_path = os.path.join(self.image_dir, image_name)
        label_path = os.path.join(self.label_dir, image_name.replace('.jpg', '.txt').replace('.png', '.txt'))

        image = Image.open(image_path).convert("RGB")
        image = self.transform(image)

        with open(label_path, 'r') as f:
            line = f.readline().strip().split()
            line = list(map(float, line))

        bbox = torch.tensor(line[1:5], dtype=torch.float32)  # cx, cy, w, h
        keypoints = torch.tensor(line[5:], dtype=torch.float32).view(21, 3)

        target = {
            "bbox": bbox,
            "keypoints": keypoints,
            "image_id": image_name
        }

        return image, target
```

#### Heatmaps

```python
# Heatmaps
import os
import torch
from torch.utils.data import Dataset
from torchvision import transforms
from PIL import Image
import numpy as np

class HandKeypointDetectionDataset(Dataset):
    def __init__(self, image_dir, label_dir, image_size=224, heatmap_size=56, sigma=1.5):
        self.image_dir = image_dir
        self.label_dir = label_dir
        self.image_files = sorted([f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))])
        self.image_size = image_size
        self.heatmap_size = heatmap_size
        self.sigma = sigma

        self.transform = transforms.Compose([
            transforms.Resize((image_size, image_size)),
            transforms.ToTensor(),
        ])

        self.valid_image_files = self._filter_valid_samples()

    def _filter_valid_samples(self):
        valid_files = []
        for image_name in self.image_files:
            label_path = os.path.join(self.label_dir, image_name.replace('.jpg', '.txt').replace('.png', '.txt'))
            try:
                with open(label_path, 'r') as f:
                    line = list(map(float, f.readline().strip().split()))
                    keypoints = torch.tensor(line[5:], dtype=torch.float32).view(21, 3)
                    if torch.all((keypoints[:, 0] >= 0) & (keypoints[:, 0] <= 1) &
                                 (keypoints[:, 1] >= 0) & (keypoints[:, 1] <= 1)):
                        valid_files.append(image_name)
            except:
                continue
        return valid_files

    def __len__(self):
        return len(self.valid_image_files)

    def __getitem__(self, idx):
        image_name = self.valid_image_files[idx]
        image_path = os.path.join(self.image_dir, image_name)
        label_path = os.path.join(self.label_dir, image_name.replace('.jpg', '.txt').replace('.png', '.txt'))

        image = Image.open(image_path).convert("RGB")
        image = self.transform(image)

        with open(label_path, 'r') as f:
            line = list(map(float, f.readline().strip().split()))

        keypoints = torch.tensor(line[5:], dtype=torch.float32).view(21, 3)

        # Generate heatmaps
        heatmaps = self._generate_heatmaps(keypoints)

        target = {
            "heatmaps": heatmaps,   # shape: [21, heatmap_size, heatmap_size]
            "image_id": image_name,
        }

        return image, target

    def _generate_heatmaps(self, keypoints):
        heatmaps = np.zeros((21, self.heatmap_size, self.heatmap_size), dtype=np.float32)

        for i, (x, y, v) in enumerate(keypoints):
            if v < 0.5:  # visibility threshold
                continue

            x_hm = int(x * self.heatmap_size)
            y_hm = int(y * self.heatmap_size)

            if 0 <= x_hm < self.heatmap_size and 0 <= y_hm < self.heatmap_size:
                heatmaps[i] = self._draw_gaussian(heatmaps[i], (x_hm, y_hm), self.sigma)

        return torch.tensor(heatmaps)

    def _draw_gaussian(self, heatmap, center, sigma):
        mu_x, mu_y = center

        x = np.arange(0, self.heatmap_size, 1, float)
        y = np.arange(0, self.heatmap_size, 1, float)[:, np.newaxis]
        g = np.exp(- ((x - mu_x) ** 2 + (y - mu_y) ** 2) / (2 * sigma ** 2))

        heatmap = np.maximum(heatmap, g)
        return heatmap
```

```python
import torch
import torch.nn as nn
import torchvision.models as models
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights


class HandKeypointModel(nn.Module):
    def __init__(self, architecture='simple', pretrained=True):
        super(HandKeypointModel, self).__init__()
        self.architecture = architecture.lower()

        if self.architecture == 'simple':
            self.backbone = nn.Sequential(
                nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1),
                nn.BatchNorm2d(32),
                nn.ReLU(),
                nn.MaxPool2d(2),

                nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
                nn.BatchNorm2d(64),
                nn.ReLU(),
                nn.MaxPool2d(2),

                nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
                nn.BatchNorm2d(128),
                nn.ReLU(),
                nn.MaxPool2d(2),

                nn.Conv2d(128, 256, kernel_size=3, stride=1, padding=1),
                nn.BatchNorm2d(256),
                nn.ReLU(),
                nn.AdaptiveAvgPool2d((1, 1))
            )
            self.feature_dim = 256

        elif self.architecture == 'resnet18':
            base = models.resnet18(pretrained=pretrained)
            layers = list(base.children())[:-1]  # remove fc layer
            self.backbone = nn.Sequential(*layers)  # output: [B, 512, 1, 1]
            self.feature_dim = 512

        elif self.architecture == 'efficientnet_b0':
            base = efficientnet_b0(weights=EfficientNet_B0_Weights.DEFAULT if pretrained else None)
            self.backbone = base.features
            self.pool = nn.AdaptiveAvgPool2d((1, 1))
            self.feature_dim = 1280
        else:
            raise ValueError(f"Unknown architecture: {self.architecture}")

        # Общая голова
        self.head = nn.Sequential(
            nn.Flatten(),
            nn.Linear(self.feature_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(512, 67)  # bbox(4) + keypoints(21*2) + visibility(21)
        )

    def forward(self, x):
        x = self.backbone(x)
        if self.architecture == 'efficientnet_b0':
            x = self.pool(x)
        x = self.head(x)

        # Разбиваем выход
        bbox = x[:, :4]
        keypoints = x[:, 4:46].view(-1, 21, 2)
        visibility = x[:, 46:].view(-1, 21)

        return {
            "bbox": bbox,
            "keypoints": keypoints,
            "visibility": visibility
        }
```

#### Heatmaps-based

```python
import torch
import torch.nn as nn
import torchvision.models as models
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights


class HeatmapHandKeypointModel(nn.Module):
    def __init__(self, num_keypoints=21, heatmap_size=64, architecture='simple', pretrained=True):
        super(HeatmapHandKeypointModel, self).__init__()
        self.num_keypoints = num_keypoints
        self.heatmap_size = heatmap_size
        self.architecture = architecture.lower()

        if self.architecture == 'simple':
            self.backbone = nn.Sequential(
                nn.Conv2d(3, 32, kernel_size=3, padding=1),
                nn.BatchNorm2d(32),
                nn.ReLU(),
                nn.MaxPool2d(2),

                nn.Conv2d(32, 64, kernel_size=3, padding=1),
                nn.BatchNorm2d(64),
                nn.ReLU(),
                nn.MaxPool2d(2),

                nn.Conv2d(64, 128, kernel_size=3, padding=1),
                nn.BatchNorm2d(128),
                nn.ReLU(),

                nn.Conv2d(128, 256, kernel_size=3, padding=1),
                nn.BatchNorm2d(256),
                nn.ReLU()
            )
            self.backbone_out_channels = 256
            self.need_upsample = False  # already outputs [B, C, 64, 64]

        elif self.architecture == 'resnet18':
            base = models.resnet18(pretrained=pretrained)
            layers = list(base.children())[:-2]  # Keep output: [B, 512, 8, 8]
            self.backbone = nn.Sequential(*layers)
            self.backbone_out_channels = 512
            self.need_upsample = True

        elif self.architecture == 'efficientnet_b0':
            base = efficientnet_b0(weights=EfficientNet_B0_Weights.DEFAULT if pretrained else None)
            self.backbone = base.features
            self.backbone_out_channels = 1280
            self.need_upsample = True
        else:
            raise ValueError(f"Unknown architecture: {self.architecture}")

        # Head that generates heatmaps
        self.head = nn.Sequential(
            nn.Conv2d(self.backbone_out_channels, num_keypoints, kernel_size=1),
            nn.Sigmoid()  # Optional, for heatmaps in [0,1] range
        )

        # If backbone output is not the same size as heatmap_size, upsample
        if self.need_upsample:
            self.upsample = nn.Upsample(size=(heatmap_size, heatmap_size), mode='bilinear', align_corners=False)
        else:
            self.upsample = nn.Identity()

    def forward(self, x):
        x = self.backbone(x)               # -> [B, C, H, W]
        x = self.head(x)                   # -> [B, num_keypoints, H, W]
        x = self.upsample(x)               # -> [B, num_keypoints, heatmap_size, heatmap_size]
        return x                           # heatmaps
```

### Loss

#### Points and bboxes

```python
class HandKeypointLoss(nn.Module):
    def __init__(self, use_visibility_loss=True, lambda_box=1.0, lambda_kpts=10.0, lambda_vis=0.1):
        super().__init__()
        self.use_visibility_loss = use_visibility_loss

        self.box_loss_fn = nn.SmoothL1Loss()
        self.kpt_loss_fn = nn.MSELoss(reduction='none')
        self.vis_loss_fn = nn.BCEWithLogitsLoss()

        self.lambda_box = lambda_box
        self.lambda_kpts = lambda_kpts
        self.lambda_vis = lambda_vis

    def forward(self, preds, targets):
        pred_boxes = preds["bbox"]          # [B, 4]
        pred_keypoints = preds["keypoints"]  # [B, 21, 2]
        pred_visibility = preds["visibility"]  # [B, 21]

        target_boxes = targets["bbox"]        # [B, 4]
        target_keypoints = targets["keypoints"][:, :, :2]  # [B, 21, 2]
        target_vis = targets["keypoints"][:, :, 2]          # [B, 21]

        # --- 1. Bounding box loss ---
        loss_bbox = self.box_loss_fn(pred_boxes, target_boxes)

        # --- 2. Keypoint regression loss ---
        # Only apply to visible keypoints
        vis_mask = (target_vis > 0).unsqueeze(-1)  # [B, 21, 1]
        keypoint_diff = self.kpt_loss_fn(pred_keypoints, target_keypoints)  # [B, 21, 2]
        keypoint_loss = (keypoint_diff * vis_mask).sum() / (vis_mask.sum() + 1e-6)

        # --- 3. Visibility prediction loss ---
        if self.use_visibility_loss:
            loss_vis = self.vis_loss_fn(pred_visibility, target_vis)
        else:
            loss_vis = torch.tensor(0.0, device=pred_boxes.device)

        # --- Final weighted loss ---
        total_loss = (
            self.lambda_box * loss_bbox +
            self.lambda_kpts * keypoint_loss +
            self.lambda_vis * loss_vis
        )

        return total_loss, {
            "bbox_loss": loss_bbox.item(),
            "keypoint_loss": keypoint_loss.item(),
            "visibility_loss": loss_vis.item() if self.use_visibility_loss else 0.0
        }
```

#### Heatmaps-based

```python
class HeatmapKeypointLoss(nn.Module):
    def __init__(self):
        super(HeatmapKeypointLoss, self).__init__()
        self.criterion = nn.MSELoss()

    def forward(self, pred_heatmaps, target_heatmaps):
        """
        Args:
            pred_heatmaps: Tensor of shape [B, 21, H, W]
            target_heatmaps: Tensor of shape [B, 21, H, W]
        """
        loss = self.criterion(pred_heatmaps, target_heatmaps)
        return loss
```

#### Heatmaps

```python
import torch
import cv2
import numpy as np
import matplotlib.pyplot as plt

def predict_and_draw(model, image_tensor, orig_image, threshold=0.1, device="cuda"):
    """
    Args:
        model: trained model
        image_tensor: [1, 3, H, W] normalized and resized
        orig_image: original image (H, W, 3) as loaded by cv2 (BGR)
        threshold: minimum heatmap peak value to draw a keypoint
        device: 'cuda' or 'cpu'
    Returns:
        vis_image_rgb: visualization image (RGB format)
    """
    model.eval()
    image_tensor = image_tensor.to(device)

    with torch.no_grad():
        pred_heatmaps = model(image_tensor)  # [1, 21, H, W]

    pred_heatmaps = pred_heatmaps.squeeze(0).cpu().numpy()  # [21, H, W]
    h_img, w_img, _ = orig_image.shape
    h_heatmap, w_heatmap = pred_heatmaps.shape[1], pred_heatmaps.shape[2]

    keypoints = []

    for i in range(pred_heatmaps.shape[0]):
        heatmap = pred_heatmaps[i]
        max_val = np.max(heatmap)
        if max_val < threshold:
            keypoints.append(None)
            continue

        y, x = np.unravel_index(np.argmax(heatmap), heatmap.shape)
        # Convert heatmap coords to original image coords
        x = int((x / w_heatmap) * w_img)
        y = int((y / h_heatmap) * h_img)
        keypoints.append((x, y))

    vis_image = orig_image.copy()

    for pt in keypoints:
        if pt is not None:
            cv2.circle(vis_image, pt, 3, (0, 0, 255), -1)

    vis_image_rgb = cv2.cvtColor(vis_image, cv2.COLOR_BGR2RGB)
    return vis_image_rgb
```
